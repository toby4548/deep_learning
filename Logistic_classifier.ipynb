```python
"""Softmax."""
import numpy as np

def softmax(x):
    """Compute softmax values for each sets of scores in x."""
    #pass  # TODO: Compute and return softmax(x)
    return  np.exp(x)/np.sum(np.exp(x), axis = 0)

# Plot softmax curves
import matplotlib.pyplot as plt

test_scores = [3.0, 1.0, 0.2]
print(softmax(test_scores))

x = np.arange(-2.0, 6.0, 0.1)
scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)])
#print(softmax(scores))
plt.plot(x, softmax(scores).T, linewidth=2)
plt.show()
```

```python
test_scores = np.array([3.0, 1.0, 0.2])
print(softmax(test_scores*10)) # multiple scores by 10
x = np.arange(-2.0, 6.0, 0.1)
scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)])
plt.plot(x, softmax(scores*10).T, linewidth=2)
plt.show()
```

if we multiple the score by 10, the result getting more close to 0 or 1 .
*******It means that the classifier is confident to its result.********

```python
test_scores = np.array([3.0, 1.0, 0.2])
print(softmax(test_scores/10)) # devide scores by 10
x = np.arange(-2.0, 6.0, 0.1)
scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)])
plt.plot(x, softmax(scores/10).T, linewidth=2)
plt.show()
```

if we devide score by 10, the probability become more close to normal distributed
*******It means that the classifier is not sure about its result.********
